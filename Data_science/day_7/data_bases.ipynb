## There is few examples of how to write type casting and viseversa

```python
df.info()
```

```python
df['height'] = df['height'].astype(int)
print(df['height'].dtype) 
```

```python
df['weight'] = df['weight'].astype(int)
print(df['weight'].dtype)  
```

```python
df['favourite_Dish'] = df['favourite_Dish'].astype(str)
print(df['favourite_Dish'].dtype)  
```

```python
df['favourite_Dish'] = df['favourite_Dish'].astype('category')
print(df['favourite_Dish'].dtype)  # Output: category
```
