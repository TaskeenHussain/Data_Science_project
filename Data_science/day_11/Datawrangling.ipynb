# Box-and-Whisker Plot
A box-and-whisker plot, often called a box plot, is a graphical representation of a dataset that shows its spread and identifies potential outliers. It provides a summary of the following statistics:

1-Minimum (excluding outliers)
2-First Quartile (Q1): 25th percentile
3-Median (Q2): 50th percentile
4-Third Quartile (Q3): 75th percentile
Maximum (excluding outliers)
## Key Components
1-Box:
The box spans from Q1 (25th percentile) to Q3 (75th percentile).
The line inside the box represents the median (Q2).\
2-Whiskers:
Extend from the edges of the box to the smallest and largest values within the allowable range.
Allowable range is determined using the Interquartile Range (IQR).\
3-Outliers:
Data points beyond the whiskers are considered outliers.\
# Interquartile Range (IQR) Formula
The IQR measures the spread of the middle 50% of the data.\

IQR = Q3 - Q1

IQR=Q3âˆ’Q1

# Detecting Outliers Using IQR
 To identify outliers, use the following rules:

### Lower Bound:ð‘„1âˆ’1.5Ã—IQR,
Q1âˆ’1.5Ã—IQR
### Upper Bound:ð‘„3+1.5Ã—IQR
Q3+1.5Ã—IQR\
Data points outside these bounds are considered outliers.

## Interpretation
The box shows where most of the data lies (middle 50%).
The whiskers indicate the spread of the data, excluding outliers.
Points beyond the whiskers are flagged as outliers.

```python
#Example in Python\
#python
# Copy code
import matplotlib.pyplot as plt
import numpy as np

# Example dataset
data = [7, 8, 9, 10, 15, 22, 30, 35, 40, 50, 100]  # 100 is an outlier

# Create a box plot
plt.boxplot(data, vert=False)
plt.title("Box-and-Whisker Plot")
plt.xlabel("Values")
plt.show()

# Calculate IQR and detect outliers
Q1 = np.percentile(data, 25)  # First quartile
Q3 = np.percentile(data, 75)  # Third quartile
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]
print(f"Outliers: {outliers}")
```
