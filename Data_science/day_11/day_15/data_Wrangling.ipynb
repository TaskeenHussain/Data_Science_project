# Data Wrangling
Data wrangling is process of transforming and cleaning and organizing data into a format that is suitable for  analysis and visualization . It involves several steps: 

1.  **Data Cleaning**: This step involves removing any missing or duplicate data, handling outliers, and correcting any errors in the data. 
2.  **Data Transformation**: This step involves converting the data into a suitable format for analysis. This may involve aggregating data, creating new variables, or transforming existing variables. 
3. **Data Integration**: This step involves combining data from multiple sources into a single dataset.       
### Data wrangling helps us to ensure that our data is accurate, complete, and consistent, which is essential for making informed decisions

```python
# IQR Method 
Q1 = df['age'].quantile(.25)
 
Q3 = df['age'].quantile(.75)
IQR = Q3 - Q1- Q1
IQR
```

```python
# IQR Method 
Q1 = df['age'].quantile(.25)
Q3 = df['age'].quantile(.75)
IQR = Q3 - Q1
IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['age']>lower_bound) & (df['age']<upper_bound)]
```
