```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import kahsti datset
kashti = sns.load_dataset("titanic")
kashti.head()
```

```python
# To chk null values
kashti.isnull()
```

```python
# How to change the titles of X and Y axis ?
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True,
                  palette={1:"seagreen", 0:"grey"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="who", margin_titles=True,
                  palette={1:"seagreen", 0:"grey"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="embark_town", margin_titles=True,
                  palette={1:"seagreen", 0:"grey"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure only numeric data is used for the correlation
numeric_kashti = kashti.select_dtypes(include=['number'])

# Compute the correlation matrix
corr = numeric_kashti.corr()

# Plot the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr) #, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure only numeric data is used for the correlation
numeric_kashti = kashti.select_dtypes(include=['number'])

# Compute the correlation matrix
corr = numeric_kashti.corr()

# Plot the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar=True)

plt.title("Correlation Heatmap")
plt.show()
```
